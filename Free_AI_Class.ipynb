```python
# ============================================
# TASK 1: Two Sum Function
# ============================================

def two_sum(a: int, b: int) -> int:
    """
    Calculates the sum of two integers.

    Parameters:
    a (int): First integer
    b (int): Second integer

    Returns:
    int: Sum of a and b
    """
    return a + b

# Test the function
print("Testing two_sum function:")
print(f"two_sum(5, 3) = {two_sum(5, 3)}")    # Expected: 8
print(f"two_sum(-2, 7) = {two_sum(-2, 7)}")  # Expected: 5
print(f"two_sum(0, 0) = {two_sum(0, 0)}")    # Expected: 0


# ============================================
# TASK 2: Prime Number Checker
# ============================================

def is_prime(n: int) -> bool:
    """
    Checks if a number is prime.

    Parameters:
    n (int): Number to check

    Returns:
    bool: True if prime, False otherwise
    """
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    # Check odd divisors up to square root
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Test with if-else statement as requested
test_numbers = [1, 2, 4, 7, 10, 13, 15, 17]

print("\n" + "="*50)
print("Prime Number Checker Results:")
print("="*50)

for num in test_numbers:
    if is_prime(num):
        print(f"{num} → This is a prime number")
    else:
        print(f"{num} → This is an Ordinary number")
```

```python
# importing necessary libraries and setting style for better visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# LOADING THE DATASET
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

try:
    df = pd.read_csv(url)
    print("✅ Dataset loaded successfully!")
    print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
except Exception as e:
    print(f"❌ Error loading dataset: {e}")
```

```python
# INITIAL DATA INSPECTION CHECKING THE FIRST AND THE LAST 5 ROWS

print("\n" + "="*60)
print("INITIAL DATA INSPECTION")
print("="*60)

# First 5 rows
print("\n📋 First 5 rows:")
print(df.head())

# Last 5 rows
print("\n📋 Last 5 rows:")
print(df.tail())
```

```python
# Checking the Column names and data types with Basic Statistics
print("\n📊 Column Information:")
print(df.info())

# Basic statistics
print("\n📈 Statistical Summary (Numerical Columns):")
print(df.describe())
```

```python
# HANDLING MISSING VALUES

print("\n" + "="*60)
print("MISSING VALUES ANALYSIS")
print("="*60)

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Percentage (%)': missing_percentage
})
print(missing_df[missing_df['Missing Count'] > 0].sort_values('Percentage (%)', ascending=False))
```

```python
# Visualizing missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#  EXPLORING TARGET VARIABLE -THE SURVIVAL

print("\n" + "="*60)
print("SURVIVAL ANALYSIS")
print("="*60)

survival_counts = df['Survived'].value_counts()
print(f"\nSurvival distribution:\n{survival_counts}")
print(f"Survival rate: {df['Survived'].mean()*100:.2f}%")

# Survival count plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
sns.countplot(data=df, x='Survived', ax=axes[0], palette=['#ff7f0e', '#1f77b4'])
axes[0].set_title('Passenger Count by Survival Status', fontsize=14)
axes[0].set_xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].set_xticklabels(['Did Not Survive', 'Survived'])

# Pie chart
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90,
                                         colors=['#ff7f0e', '#1f77b4'], ax=axes[1])
axes[1].set_title('Survival Percentage', fontsize=14)
axes[1].set_ylabel('')
axes[1].legend(['Did Not Survive', 'Survived'], loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# ANALYZING THE CATEGORICAL FEATURES

print("\n" + "="*60)
print("CATEGORICAL FEATURES ANALYSIS")
print("="*60)

categorical_cols = ['Sex', 'Pclass', 'Embarked', 'SibSp', 'Parch']

for col in categorical_cols:
    if col in df.columns:
        print(f"\n {col} distribution:")
        print(df[col].value_counts(dropna=False))

        # Survival rate by category
        survival_by_cat = df.groupby(col)['Survived'].mean().sort_values(ascending=False)
        print(f"\n   Survival rate by {col}:")
        print(survival_by_cat)
```

```python
# Visualization: Survival by Gender and Passenger Class
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Survival by Gender
sns.barplot(data=df, x='Sex', y='Survived', ax=axes[0], palette=['#ff7f0e', '#1f77b4'])
axes[0].set_title('Survival Rate by Gender', fontsize=14)
axes[0].set_ylabel('Survival Rate', fontsize=12)
axes[0].set_ylim(0, 1)

# Survival by Passenger Class
sns.barplot(data=df, x='Pclass', y='Survived', ax=axes[1], palette='viridis')
axes[1].set_title('Survival Rate by Passenger Class', fontsize=14)
axes[1].set_ylabel('Survival Rate', fontsize=12)
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
# ANALYZING THE NUMERICAL FEATURES

print("\n" + "="*60)
print("NUMERICAL FEATURES ANALYSIS")
print("="*60)

numerical_cols = ['Age', 'Fare']

for col in numerical_cols:
    print(f"\n {col} statistics:")
    print(f"   Min: {df[col].min():.2f}")
    print(f"   Max: {df[col].max():.2f}")
    print(f"   Mean: {df[col].mean():.2f}")
    print(f"   Median: {df[col].median():.2f}")
    print(f"   Std Dev: {df[col].std():.2f}")
```

```python
# Visualization: Age and Fare distribution by Survival
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age distribution
sns.histplot(data=df, x='Age', hue='Survived', bins=30, alpha=0.6, ax=axes[0,0])
axes[0,0].set_title('Age Distribution by Survival', fontsize=14)
axes[0,0].set_xlabel('Age', fontsize=12)


# Fare distribution (log scale for better visualization)
sns.histplot(data=df, x='Fare', hue='Survived', bins=50, alpha=0.6, ax=axes[0,1])
axes[0,1].set_title('Fare Distribution by Survival', fontsize=14)
axes[0,1].set_xlabel('Fare', fontsize=12)
axes[0,1].set_xscale('log')

# Box plot - Age vs Survival
sns.boxplot(data=df, x='Survived', y='Age', ax=axes[1,0], palette=['#ff7f0e', '#1f77b4'])
axes[1,0].set_title('Age Distribution by Survival Status', fontsize=14)
axes[1,0].set_xticklabels(['Did Not Survive', 'Survived'])

# Box plot - Fare vs Survival
sns.boxplot(data=df, x='Survived', y='Fare', ax=axes[1,1], palette=['#ff7f0e', '#1f77b4'])
axes[1,1].set_title('Fare Distribution by Survival Status', fontsize=14)
axes[1,1].set_xticklabels(['Did Not Survive', 'Survived'])

plt.tight_layout()
plt.show()
```

```python
# CORRELATION ANALYSIS

print("\n" + "="*60)
print("CORRELATION ANALYSIS")
print("="*60)

# Select numerical columns for correlation
numerical_df = df.select_dtypes(include=[np.number])

# Correlation matrix
correlation_matrix = numerical_df.corr()
print("\nCorrelation with Survival (Target):")
print(correlation_matrix['Survived'].sort_values(ascending=False))

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, linewidths=1)
plt.title('Feature Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()
```
